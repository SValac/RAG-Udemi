# Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

## Method 1: Using Docx2txtLoader

```python
print("1 Using Docx2txtLoader:")
try:
  docx_loader = Docx2txtLoader("data/word_files/proposal.docx")
  docs = docx_loader.load()
  print(f"Loaded {len(docs)} documents with Docx2txtLoader")
  print(f"Content preview: {docs[0].page_content[:200]}...")
  print(f"Metadata: {docs[0].metadata}")
except Exception as e:
  print(f"Error loading DOCX with Docx2txtLoader: {e}")
```

## Method 2: Using UnstructuredWordDocumentLoader


```python
print("2 Using UnstructuredWordDocumentLoader:")
try:
  unstructured_loader =UnstructuredWordDocumentLoader("data/word_files/proposal.docx", mode="elements")
  unstructured_docs = unstructured_loader.load()
  print(f"Loaded {len(unstructured_docs)} elements with UnstructuredWordDocumentLoader")
  for i, doc in enumerate(unstructured_docs[:3]):
    print(f"\nElement {i+1}")
    print(f"Metadata: {doc.metadata}")
    print(f"Content preview: {doc.page_content[:200]}...")

except Exception as e:
  print(f"Error loading DOCX with UnstructuredWordDocumentLoader: {e}")
```
